```python
import tensorflow as tf
from  tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
#tf.logging.set_verbosity(tf.logging.ERROR)

num_classes=5

inputs_ = tf.placeholder(tf.float32, shape=(None, 299, 299))
#print(inputs_.shape)

#MODEL LAYERS inputs_>>logits

x = tf.reshape(inputs_, [-1, 299, 299, 1])  
x = Conv2D(32, (3, 3),padding='same', name='layer_conv1', activation=tf.nn.relu)(x)
x = MaxPooling2D(pool_size=2, strides=2)(x)
x = Conv2D(32, (3, 3),padding='same', name='layer_conv1', activation=tf.nn.relu)(x)
x = MaxPooling2D(pool_size=2, strides=2)(x)
x = Flatten()(x)
x = Dense(name='layer_fc1',units=128, activation=tf.nn.relu)(x)  # tf.layers.dense (N,Input_shape)--> (N, Output_shape)  https://www.tensorflow.org/api_docs/python/tf/layers/dense
x = Dropout(rate=0.5, noise_shape=None, seed=None)(x)
x = Dense(name='layer_fc2',units=num_classes, activation=tf.nn.relu)(x)

logits=x


# PREDICTIONS

y_pred     = tf.nn.softmax(logits=logits)
y_pred     = tf.identity(y_pred, name="output_pred")

y_pred_cls = tf.argmax  (y_pred, axis=1)
y_pred_cls = tf.identity(y_pred_cls, name="output_cls")



print(logits.shape)
print("y_pred.shape", y_pred.shape)
print("y_pred_cls.shape", y_pred_cls.shape)
```
